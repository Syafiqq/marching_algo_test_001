```python
import numpy as np
```

```python
linespace = np.linspace(0, 7, 8).reshape((2, 2, 2))
linespace
```

```python
flatten = linespace.flatten(order='F')
flatten
```

```python
repr(flatten)
```

```python
linespace.shape
```

```python
shape = np.array(linespace.shape)
shape
```

```python
shape_1 = np.array(linespace.shape) + 1
shape_1
```

```python
import pyvista as pv
grid = pv.UniformGrid()
grid.dimensions = shape_1
grid.origin = (0, 0, 0)
grid.spacing = (2, 2, 2)
grid.cell_data["values"] = flatten
grid.cell_data["values"] = [0., 0., 0., 0., 0., 0., 0., 0.]
grid.cell_data["values"] = [0., 4., 2., 6., 1., 5., 3., 7.]
grid.cell_data["values"] = [0., 1., 2., 3., 4., 5., 6., 7.]
grid.plot(show_edges=True, show_axes=True, show_bounds=True)
```
