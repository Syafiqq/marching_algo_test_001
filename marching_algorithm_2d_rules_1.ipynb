```python
import numpy as np
import pyvista as pv
import marching_2d_rules_1 as rules
```

```python
points = np.array(
    [
        [10, 10, 00],
        [20, 10, 00],
        [10, 20, 00],
        [20, 20, 00],
    ]
    , dtype=np.float)
points
```

```python
def plot_data(cloud_points, faces):
    cloud = pv.PolyData(cloud_points, faces=faces)
    plot = pv.Plotter()
    plot.add_points(points)
    plot.add_mesh(cloud, show_edges=True, edge_color='r')
    plot.show_bounds()
    plot.show_grid()
    plot.camera_position = 'xy'
    plot.show()
```

```python
def mid_point(p1, p2):
    pt = abs(p1 - p2) / 2
    return p1 + pt
```

```python
point = np.array([10, 10, 0])
point_2 = np.array([20, 10, 0])
mid_point(point, point_2)
```

```python
# Rule 2
# 2 . 3
# . . .
# 0 . 1

# . . .
# 2 . .
# 0 1 .
rule_type = 0b1000
points_r = rules.to_points(points, rule_type)
faces_r = rules.to_faces(np.arange(0, len(points_r), 1), rule_type)

plot_data(points_r, faces=faces_r)
```
