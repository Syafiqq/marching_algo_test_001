```python
import numpy as np
import pyvista as pv
```

```python
points = np.array(
    [
        [10, 10, 00],
        [20, 10, 00],
        [10, 20, 00],
        [20, 20, 00],
    ]
    , dtype=np.float)
points
```

```python
def mid_point(p1, p2):
    pt = abs(p1 - p2) / 2
    return p1 + pt
```

```python
point = np.array([10, 10, 0])
point_2 = np.array([20, 10, 0])
mid_point(point, point_2)
```

```python
# Rule 1
# Empty
```

```python
# Rule 2
#p 2 . 3
#p . . .
#p 0 . 1

#f . . .
#f 2 . .
#f 0 1 .
p0 = points[0]
p1 = mid_point(points[0], points[1])
p2 = mid_point(points[0], points[2])

points_rule_2 = [p0, p1, p2]
points_rule_2_faces = [4, 0, 1, 2, 0]

cloud = pv.PolyData(points_rule_2, faces=points_rule_2_faces)
plot = pv.Plotter()
plot.add_points(points)
plot.add_mesh(cloud, show_edges=True, edge_color='r')
plot.show_bounds()
plot.show_grid()
plot.camera_position = 'xy'
plot.show()
```
