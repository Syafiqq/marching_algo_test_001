```python
import numpy as np
import pyvista as pv
import marching_3d_rules_minimal as rules
```

```python
points = np.array(
    [
        [10, 10, 10],
        [20, 10, 10],
        [10, 20, 10],
        [20, 20, 10],
        [10, 10, 20],
        [20, 10, 20],
        [10, 20, 20],
        [20, 20, 20],
    ]
    , dtype=np.float)
points
```

```python
def plot_data(cloud_points, faces):
    cloud = pv.PolyData(cloud_points, faces=faces)
    plot = pv.Plotter()
    plot.add_points(points)
    plot.add_mesh(cloud, show_edges=True, edge_color='r')
    plot.show_bounds()
    plot.show_grid()
    plot.show()
```

```python
# Rule 51
# 1 0 0 1   0 1 0 0
#
#   6     7
# 4     o
#   2     o
# o     1
#
rule_type = 0b10010100
points_r = rules.to_points(points, rule_type)
faces_r = rules.to_faces(np.arange(0, len(points_r), 1), rule_type)

plot_data(points_r, faces=faces_r)
```
