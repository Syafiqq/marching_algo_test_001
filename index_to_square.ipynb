```python
import numpy as np
import pyvista as pv
```

```python
nxy = 4
dims = np.array([nxy, nxy, 1])
dims
```

```python
dims_f = np.prod(dims)
dims_f
```

```python
dims_1 = [dims[0] + 1, dims[1] + 1, dims[2]]
dims_1
```

```python
grid = pv.UniformGrid(
    dims=dims_1,
    spacing=(1, 1, 1),
    origin=(0, 0, 0),
)
```

```python
points_t = np.transpose(grid.points.T)
points_t
```

```python
cloud = pv.PolyData(points_t)
plot = pv.Plotter()
plot.add_points(points_t, render_points_as_spheres=True, point_size=20.0)
plot.show_bounds()
plot.show_grid()
plot.camera_position = 'xy'
plot.show()
```

Try To Loop the value

```python
cxy = dims_1[0] * dims_1[1]
cxy
```

```python
cx = dims_1[0]
cx
```

```python
for dz in np.arange(0, dims[2], 1, dtype=np.int):
    for dy in np.arange(0, dims[1], 1, dtype=np.int):
        for dx in np.arange(0, dims[0], 1, dtype=np.int):
            pt = dz * cxy + dy * cx + dx * 1
            p0 = pt
            p1 = pt + 1
            p2 = p0 + cx
            p3 = p1 + cx
            print([points_t[p0], points_t[p1], points_t[p2], points_t[p3]])
```

Try To Loop the value 1

```python
cxy_f = dims[0] * dims[1]
cx_f = dims[0]

for i in np.arange(0, dims_f, 1, dtype=np.int):
    dz = i // cxy_f                         # i div (dimension x and y) to get axis z
    dz_r = i - (dz * cxy_f)                 # get the remaining of dz
    dy = dz_r // cx_f                       # remain of dz div (dimension of x) to get axis y
    dy_r = dz_r - (dy * cx_f)               # get the remaining of dy
    dx = dy_r                               # get the dx

    pt = dz * cxy + dy * cx + dx * 1        # mapping value to match array of points
    p0 = pt                                 # find the lowest x
    p1 = pt + 1                             # find the lowest x + 1
    p2 = p0 + cx                            # find the lowest x with y + 1
    p3 = p1 + cx                            # find the lowest x + 1 with y + 1
    print([points_t[p0], points_t[p1], points_t[p2], points_t[p3]])
```
